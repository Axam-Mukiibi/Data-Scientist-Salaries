```python
ds2.drop(ds2[['experience_level','employment_type','company_size']], axis=1, inplace=True)
```

```python
ds3 = ds2.drop(ds2[['job_title', 'salary_currency', 'employee_residence', 'company_location']], axis=1)
```

```python
ds3
```

```python
#training data
from sklearn.model_selection import train_test_split
```

```python
X = ds3.drop('salary_in_usd', axis=1).values
```

```python
y = ds3['salary_in_usd'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#normlaizing data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#Creating Model
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(14,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(14,activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=600)
```

```python
#training history
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
#model evaluation
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
model.evaluate(X_train, y_train, verbose=0)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
predictions = pd.Series(predictions.reshape(201,))
```

```python
pred_df = pd.DataFrame(y_test, columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df, predictions], axis=1)
```

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
```
